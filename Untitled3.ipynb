```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import networkx as 
csv_data = pd.read_csv("acci.csv",engine = 'python')
csv_data
```

```python
group_names = ['작다','보통','크다']
csv_data['사고 빈도수'],mybin = pd.cut(csv_data['무인'],3,labels=group_names,retbins=True)
csv_data
```

```python
csv_data = pd.concat([csv_data, pd.get_dummies(csv_data['사고 빈도수'])],axis=1,sort=False)
csv_data
```

```python
csv_data['사고 빈도수'] = (csv_data['사고 빈도수']-csv_data['사고 빈도수'].min())/(csv_data['P'].max()- csv_data['P'].min())
csv_data
```
